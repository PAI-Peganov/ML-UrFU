## Проблемы обработки последовательностей

Сегодня мы познакомимся с обработкой последовательностей на примере текста. Конечно, последовательностью может быть музыка или видео, временные ряды или траектория движения робота, последовательности аминокислот в белке или много чего ещё, однако на момент 2026 года самая богатая область DS, связанная с последовательностями, это всё-таки NLP

Natural Language Processing (обработка естественного языка) включает в себя много различных задач, среди них можно выделить:

1. Классификация документов
2. Определение спама
3. Определение частей речи
4. Исправление орфографических ошибок и опечаток
5. Анализ тональности текста
6. Summary
7. Машинный перевод

И многое-многое другое

1. **Many-to-one**. На вход подается последовательность объектов, на выходе один объект. Пример : классификация текстов или видео. 
2. **One-to-many**. На вход подается один объект, на выходе последовательность объектов. Пример: генерация заголовка к изображению (image captioning).
3. **Many-to-many**. На входе и выходе последовательности нефиксированной длины. Примеры: машинный перевод, суммаризация текста, генерация заголовка к статье.
4. **Синхронизированный вариант many-to-many**. На входе и выходе последовательности одинаковой длины, токены входной явно сопоставлены соответствующим токенам выходной. Пример: генерация покадровых субтитров к видео, PoS-tagging (part of speech tagging, для каждого слова в предложении предсказываем, что это за часть речи).

Можно выделить несколько глобальный проблем при работе с текстом

1. Мы не можем просто так взять и отправить текст, как некоторый входной вектор $x$ в модель
2. Текст - это последовательность
3. Текст состоит из дискретных сущностей
4. Не заданы операции $+$ и $\times$ над словами..

### TF-IDF (term frequency inverse document frequency)

Будем называть один сэмпл со словами - **документом**. А одно слово (токен) в тексте **term**. 

Тогда вводятся 2 такие величины:

$$TF(t, d) = \frac{N_t}{N_d}$$

$$IDF(t) = \log{\frac{N}{count(d \in D| t \in d)}}$$

**TF** (term frequency — частота слова) — отношение числа вхождений некоторого слова к общему числу слов документа. Таким образом, оценивается важность слова $t_i$ в пределах отдельного документа $d$

**IDF** (inverse document frequency — обратная частота документа) — инверсия частоты, с которой некоторое слово встречается в документах коллекции. 

Таким образом, мера TF-IDF является произведением двух сомножителей:

$$tf-\hspace{0pt}idf(t, d, D) = tf(t, d) \times idf(t, D)$$

Большой вес в TF-IDF получат слова с высокой частотой в пределах конкретного документа и с низкой частотой употреблений в других документах.

### Идея Word2Vec



Одним из воплощений подхода, описанного выше, является Word2vec. Впервые он был предложен Т.Миколовым в 2013 году в [статье](https://arxiv.org/abs/1301.3781) Efficient Estimation of Word Representations in Vector Space.

**Что мы хотим:**

Мы хотим выучить векторы слов **небольшой** размерности, котрые отражали бы смысл слов: слова можно было бы сравнивать между собой в привычном векторном смысле (в векторном пространстве)

Такие выученные векторы мы будем называть `embeddings` (плотные векторные представления)

Мы будем учить нейронную сеть по слову предсказывать контекст этого слова (и наоборот)

Пусть есть набор текстов. Будем идти по этому набору окном размера $k$, и в каждом положении окна по центральному слову будем предсказывать слова, находящиеся в текущем окне.

![alt](../data/w2v1.png)

![alt](../data/w2v2.png)


Будем решать некоторую proxy-задачу по восстановлению зависимости: по входному слову хотим оценить распределение вероятностей других слов из словаря. Будем делать всё это итеративно:

![alt](../data/word2vec.png)

В архитектуре Skip-gram мы решаем задачу классификации: по заданному центральному слову $w_t$ необходимо предсказать вероятные контекстные слова $w_{t+j}$, находящиеся в рамках фиксированного окна $m$.

Цель обучения — максимизировать вероятность наблюдения контекстных слов для всего корпуса текста:

$$
\frac{1}{T} \sum_{t=1}^{T} \sum_{-m \le j \le m, j \neq 0} \log P(w_{t+j} | w_t; \theta)
$$

Где $T$ — длина текста, $m$ — радиус окна, а $\theta$ — параметры модели.




![alt](../data/mEGe8.png)


Модель представляет собой полносвязную сеть с одним скрытым слоем.

- **Вход (Input)**: Вектор $x_w$, представляющий слово $w$ в формате One-hot encoding. Размерность вектора $V$ (размер словаря).
- **Скрытый слой (Hidden)**: Линейный слой без функции активации. Веса этого слоя — матрица $W \in \mathbb{R}^{V \times d}$, где $d$ — размерность эмбеддинга (например, 300).
- **Выходной слой (Output)**: Матрица контекстных весов $W' \in \mathbb{R}^{d \times V}$. На выходе применяется функция Softmax.

$$
P(c|w) = \frac{\exp(u_c)}{\sum_{j=1}^{V} \exp(u_j)}
$$

$u_j = <W_i, W_j>$ - логит между центральным словом $w$ ($i$-го из словаря) и контекстным словом $j$


![alt](../data/cbow_skip.png)

На самом деле мы теперь можем перейти от решения задачи мультиклассификации к решению задачи классификации. Давайте будем решать такую задачу: встречается \ не встречается ли заданное слово с другим в контексте.

На каждом шаге обучения модель учится отличать:

* Реальную пару $(w, c)$ - действительно встретившуюся в тексте
* Шумовую пару $(w, c_i)$ - где $c_i$ случайное слово из словаря и не из контекста.

## RNN

Итак, мы представили текст в виде последовательности векторов, соответствующих словам или их кусочкам. Как с ней работать? Один из вариантов мы уже рассматривали: можно посмотреть на последовательность из $k$ векторов размерности $d$ как на «изображение» $k \times 1$ с $d$ каналами. После чего использовать уже знакомые нам свёрточные нейросети, только с одномерными свёртками вместо двумерных.

В каких-то случаях это действительно будет работать, но всё же есть несколько сомнительных моментов:

* Хотя изображения тоже могут быть разного размера, всё же в датасете редко попадаются рядом картинки  $1920 \times 1080$ и $3 \times 3 $, а среди, скажем, отзывов на ресторан могут попадаться как труды, сопоставимые по размеру с «Войной и миром», так и безликие «Да, вроде норм». И если обработать слишком длинное предложение нам поможет (с потерей информации, конечно) global pooling, слишком короткое может что-нибудь поломать, особенно если мы забываем про padding.
* Слегка философское соображение. Изображение однородно, в нём нет предпочтительных направлений, тогда как текст пишется и читается последовательно. Нам может показаться, что это стоит использовать: при обработке очередного токена обращаться к предыдущим, как к его контексту.


В последнем соображении уже непосредственно видна идея рекуррентных нейронных сетей (recurrent networks, RNN):

![alt](../data/rnn.svg)


Давайте разберёмся, что тут происходит. Чтобы хранить информацию о предыдущих токенах, мы вводим понятие внутренней памяти или скрытого состояния (hidden state, векторы $h_n$). В простейшем случае оно выражается одним вектором фиксированной размерности. На каждом (дискретном) шаге в сеть подаются данные (например, эмбеддинг токена), при этом происходит обновление скрытого состояния.

Пример:

$$h_n = tanh(h_{n-1}W_1 + x_nW2)$$

после чего по скрытому состоянию предсказывается выходной сигнал, к примеру, следующим образом:

$$y_n = h_nW3$$


Модификации RNN, которые помогают лучше моделировать долгосрочные зависимости (LSTM, GRU) — важная веха развития нейросетей в NLP. Следующий большой этап в развитии — механизм внимания — мы рассмотрим в следующий раз.
