## Повторение

### Обучение с учителем

$X$ — пространство объектов  
$Y$ — множество ответов (предсказаний / оценок / прогнозов)  
$y(x)$, $y: X \to Y$ — неизвестная зависимость (target function)  

Правильные ответы для каждого объекта обучающей выборки заранее известны.


### Деревья решений

Дерево **в нашем понимании** - связный **ориентированный** ациклический граф, причем в каждую вершину входит не более одного ребра.

![alt](../data/1.png)

Обычно на практике рассматривают `бинарные деревья решений` (binary decision tree)

`Бинарным деревом решений` называют дерево решений, у которого каждый разделяющий узел имеет ровно два потомка.

$h(\textbf{x}, w_i): \mathbb{R}^D \times W \rightarrow \{0, 1\}$,

$w_i \in W$ - параметры разделяющей функции

---

Функция выбора характеристик (Feature Selector Function, FSF) - это функция, получающая на вход признаковое описание объекта $\textbf{x} \in X$ и возвращающая вектор $\dot{\mathbf{x}}$ меньшей размерности, содержащий подмножество компонент исходного вектора $\mathbf{x}$

$\psi(\mathbf{x}) : \mathbb{R}^D \rightarrow \mathbb{R}^{\dot{D}}$, причём, как правило, $\dot{D}<< D $

`Важно:` Функция $\psi$ часто является персонализированной для конкретного узла

---

**Axis-aligned hyperplane weak learner**

![alt](../data/2.png)

$h(\mathbf{x}, w_i) = [t_1 > \psi(\mathbf{x}) > t_2)]$,

$w_i = (\psi, t_1, t_2)$, $\psi(\mathbf{x}) = x_j$


**General oriented hyperplane**

![alt](../data/3.png)

$h(\mathbf{x}, w_i) = [t_1 > \psi(\mathbf{x})^\top \mathbf{w} > t_2)]$,

$w_i = (\psi,\mathbf{w}, t_1, t_2)$

В данном случае $\dot{D} \geq 2$

**Quadratic surface (conic in 2D)**

![alt](../data/4.png)

$h(\mathbf{x}, w_i) = [t_1 > \phi(\psi(\mathbf{x}))^\top \mathbf{w} > t_2)]$,

$w_i = (\psi,\mathbf{w}, t_1, t_2)$

### Прирост информации (Information Gain)

Это функционал качества, который мы максимизируем при каждом разбиении:

$$IG(U, j, t) = \mathbb{H}(U) - \frac{|U_{\text{left}}|}{|U|} \mathbb{H}(U_{\text{left}}) - \frac{|U_{\text{right}}|}{|U|} \mathbb{H}(U_{\text{right}})$$


`Когда создать терминальный узел?`

`Stopping Criteria`

Терминальный узел создаётся, если выполнено одно из условий:

* Достигнута глубина дерева $d$
* $N_i$ < n
* $\mathbb{H}(S_i) < \varepsilon$


`Как создать терминальный узел?`

$$t = \argmax_kN^{(k)}_i$$

$$t = (\frac{N_i^{(1)}}{N_i}, \frac{N_i^{(2)}}{N_i}, ..., \frac{N_i^{(K)}}{N_i})$$

Далее, будем считать, что любое дерево $T$ возвращает в качестве ответа именно такой вектор вероятностей $t$.

### Идея ансамбля (композиции) алгоритмов

Пусть $a_1, a_2, ..., a_n$ - обученные алгоритмы, пусть $\mathbf{x}$ признаковое описание объекта

`Классификация`: относим $\mathbf{x}$ к классу, за который проголосовало большинство из $a_1(\mathbf{x}), a_2(\mathbf{x}), ..., a_n(\mathbf{x})$

`Регрессия`: ответом является среднее значение $a_1(\mathbf{x}), a_2(\mathbf{x}), ..., a_n(\mathbf{x})$


**Ансамблем** (**Ensemble**) назывется алгоритм, который состоит из нескольких алгоритмов МО, а процесс построения ансамбля называется **ансамблированием**.

Ответ ансамбля записывают так:

$$a(\mathbf{x}) = G(a_1(\mathbf{x}), a_2(\mathbf{x}), ..., a_n(\mathbf{x}))$$

$G$ - некоторая агрегирующая функция

При этом, каждый алгоритм $a_i$ принадлежит одному и тому же множеству алгоритмов.

Если не все алгоритмы $a_i$ принадлежат одноиу и тому же множеству алгоритмов, то тогда говорят о `fusion`, а не об `ensemble`.


## Случайный лес

`Random Forest` - это как раз ансамбль деревьев решений. То есть, в основе `RF` лежит много `DT`.

Как же объединить все деревья решений в один алгоритм?

`Дано:`

$$T_1, T_2, ..., T_n$$

`Задача`:

Построить более сильный классификатор:

$$RF(\mathbf{x}) = RF(T_1(\mathbf{x}), T_2(\mathbf{x}), ..., T_n(\mathbf{x}))$$


Считаем, что каждое дерево возращает $K$ - мерный вектор:
$$ T(\mathbf{x})= (\frac{N_i^{(1)}}{N_i}, \frac{N_i^{(2)}}{N_i}, ..., \frac{N_i^{(K)}}{N_i}), \mathbf{x} \in U$$

Тогда $RF$ в качестве агрегирующей функции ансамбля определяется следующим образом:

$$RF(\mathbf{x}) = \frac{1}{n}\sum_{i=1}^{n}T_i(\mathbf{x})$$

`Почему вообще ансамбли работают?`

`Утверждение:`

Если рассматривать значения базовых алгоритмов $a_i$ на объекте $\mathbf{x}$ как н.с.в $\xi_1, ..., \xi_n$ с одинаковым мат.ожид $\mathbb{E}[\xi_i]=y$ и одинаковой конечной дисперсией $\mathbb{D}[\xi_i]=\sigma^2$, то с.в. $\xi = \frac{1}{n}\sum_{i=1}^{n}a_i(\mathbf{x})$ имеет такое же мат. ожидание $\mathbb{E}$, но меньшую дисперсию $\mathbb{D}$

`Доказательство:`

$$\xi = \frac{1}{n}\sum_{i=1}^{n}a_i(\mathbf{x}) = \frac{1}{n}\sum_{i=1}^{n}\xi_i(\mathbf{x})$$
$$\mathbb{E}[\xi] = \mathbb{E}[\frac{1}{n}\sum_{i=1}^{n}\xi_i(\mathbf{x})] = \frac{1}{n}{\sum_{i=1}^{n}\mathbb{E}[\xi_i]} = \frac{n\mathbb{E}[\xi_i]}{n} = \mathbb{E}[\xi_i] = y$$
$$\mathbb{D}[\xi] = \mathbb{D}[\frac{1}{n}\sum_{i=1}^{n}\xi_i(\mathbf{x})] = \frac{1}{n^2}{\sum_{i=1}^{n}\mathbb{D}[\xi_i]} = \frac{n\mathbb{E}[\xi_i]}{n^2} = \frac{\mathbb{D}[\xi_i]}{n} = \frac{\sigma^2}{n}$$

С точки зрения машинного обучения, меньшая дисперсия означает меньшую склонность к переобучению.


## Построение RF

Если $T_i = T_j, \forall{i, j} \Rightarrow RF(\mathbb{x}) = T_1(\mathbb{x})$, то объединение всех таких деревьев $T_i$ не даёт никаких преимуществ по сравнению с обычным деревом.

Рассмотрим два метода, каждый из которых, будет немного ухудшать исходное дерево $T_i$. (По сути, мешаем дереву получить лучший $IG$)

### Bagging

Как же сделать так, чтобы деревья были различными?
Первый подход - `bagging` (`bootstrap aggregation`)

#### Bootstrap

![alt](../data/10.png)

Теперь у нас есть случайность в "выборе выборки", а значит и каждое дерево $T_i$ не будет являться детерменированным алгоритмом, а случайные деревья теперь можно смело объединять в случайный лес:

$$RF(\mathbf{x}) = \frac{1}{n}\sum_{i=1}^{n}T_i(\mathbf{x})$$

Действительно, теперь $T_i$ не обладает частью выборки, а это напрямую ведет в `underfitting`

### Random Node Optimization

Давайте ограничим и выберем определенные параметры разделения для каждого узла в дереве из случайного подмножество значений параметров разделения. 
Для каждого узла каждого дерева $T_i$ я выбираю подходящие параметры разделения из некоторого вектора параметров $w_i$ при выборе $h(\mathbf{x}, w_i)$, причем вектор параметров $w_i$ выбирается так, чтобьы $IG$ в этом узле максимизировался.

`Как именно мы это будем делать?`

При обучении очередного узла, мы выберем некоторое значение случайных тестов (гипотез), которые алгоритм генерирует и проверяет в каждом узле.

$$\rho$$

После выбираем ровно $\rho$ различных векторов параметров $w_1, w_2, ..., w_{\rho}$. Среди всех векторов выберем такой вектор $w_j$, что $IG$ будет максимальным

**Результат:** Деревья становятся слабокоррелированными. Индивидуальная ошибка каждого дерева (его дисперсия) может даже вырасти, но за счет их уникальности дисперсия всего ансамбля (среднего значения) резко падает.

**Вопрос**: Как $\rho$ будет влиять на деревья при $\rho=1$? А при стремлении к макс. кол-ву признаков?

---

Корень из $M$ статистически обеспечивает достаточное разнообразие, при этом оставляя высокую вероятность того, что в подмножество попадет хотя бы один сильный предиктор.

* Если мы возьмем слишком мало (например, 1 или 2), вероятность того, что в случайную выборку попадет действительно информативный признак, крайне низка. Дерево станет «слепым» и бесполезным.
* Если мы возьмем слишком много (например, 80 или 90), то в каждом узле почти гарантированно окажется самый сильный («доминирующий») признак. В итоге все деревья в ансамбле станут одинаковыми, и эффект снижения дисперсии исчезнет.


Для регрессии желательно использовать $\rho = \frac{m}{3}$

Для классификации желательно использовать $\rho = \sqrt{(m)}$

### Регуляризация

В Random Forest нет единого коэффициента регуляризации (как $\lambda$ в регрессии), потому что деревья обучаются до упора. **Регуляризация** здесь — это ограничение свободы каждого отдельного дерева, чтобы оно не «зазубривало» шум.


Это самые эффективные параметры для борьбы с переобучением:
* max_`depth: Ограничение глубины. Чем меньше глубина, тем проще модель и ниже дисперсия.
* `min_samples_leaf`: Минимальное количество объектов в листе. Если в листе остается 1-2 объекта, дерево переобучено. Установка значения 5–50 заставляет дерево искать более общие закономерности.
* `max_leaf_nodes`: Прямое ограничение количества «ответов» (листьев).


в `Random Forest` увеличение количества деревьев не ведет к переобучению. Наоборот, чем больше деревьев, тем сильнее сглаживается дисперсия. Однако после определенного числа (обычно 100–500) качество перестает расти.



### Плюсы RF


* `Устойчивость к переобучению`: Благодаря усреднению множества деревьев и бэггингу, RF крайне редко переобучается (если деревьев достаточно).
* `Не требует подготовки данных`: Вам не нужно масштабировать признаки (StandardScaler не нужен) или обрабатывать выбросы — деревьям всё равно на масштаб.
* `Работает с пропусками`: Алгоритм неплохо справляется с пропусками в данных (NaN).
* `Оценка значимости признаков (Feature Importance)`: Из коробки показывает, какие параметры больше всего влияют на результат.
* `Параллелизм`: Деревья строятся независимо друг от друга, поэтому обучение легко раскидать на все ядра процессора.
